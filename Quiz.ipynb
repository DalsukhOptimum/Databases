# [dbo].[CustomerRegister]

```sql
/****** Object:  StoredProcedure [dbo].[CustomerRegister]    Script Date: 09-04-2024 16:41:38 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create proc [dbo].[CustomerRegister]
@Name varchar(200),
@phone bigint,
@Email varchar(30),
@Adress varchar(500),
@CreatedTime datetime,
@UpdatedTime datetime,
@OperationType varchar(200),
@Exist bigint
as
begin try
 if(@OperationType = 'CustomerRegister')
 begin
 insert into Customer values(@Name,@phone,@Email,@Adress,GETDATE(),null,1);
 end;

 if(@OperationType = 'CustomerExist')
 begin
 select @Exist = COUNT(*) from Customer where Name = @Name ;
 end 

 end try 
 begin catch
 PRINT ERROR_MESSAGE()
 END catch
GO
```

# [dbo].[ShowAllQuiz]

```sql
/****** Object:  StoredProcedure [dbo].[ShowAllQuiz]    Script Date: 09-04-2024 16:41:38 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE proc [dbo].[ShowAllQuiz]
@QuizName varchar(100),
@OperationType varchar(80),
@UserScore int,
@MaxScore int out
as
begin try
if(@OperationType = 'SeeQuizNames')
begin
select * from QuizName ;
end

if(@OperationType = 'SeeQuestions')
begin
declare @Query NVARCHAR(MAX);
set @Query  = 'select * from '+ @QuizName ;
EXEC sp_executesql @Query;
end;

if(@OperationType = 'SeeMaxScore')
begin
select @MaxScore = Score from MaxScore ;
end;

if(@OperationType = 'updateScore')
begin
update MaxScore set Score = @UserScore where Id = 1 ;
end;

end try 
begin catch
print ERROR_MESSAGE();
END CATCH ;	
GO
```
