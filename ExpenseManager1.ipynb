```sql
/****** Object:  StoredProcedure [dbo].[SP_AddExpense]    Script Date: 09-04-2024 16:38:07 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE proc [dbo].[SP_AddExpense] --'salery','C',5000,'virat@gmail.com','2024-12-20 16:28:08.900'
@Comment varchar(100),
@CreditOrDebit varchar(100),
@Amount bigint ,
@Email varchar(320)

as 
begin 

declare @CreditAmount bigint = 0 ;
declare @DebitAmount bigint = 0 ;
declare @Balance bigint;
declare @Id bigint ;

select @Id = UsersId from Users where Email = @Email ;

if @Id IS NULL 
begin
select 'You Are Not Registered' as Resultmessage
return 
end 


--here we can access the last row of that particular user and then access the balance
set @CreditAmount = ISNULL ((select SUM(Amount) from Expenses where UserId = @Id AND CreditOrDebit = 'C'),0) ;
set @DebitAmount = ISNULL ((select SUM(Amount) from Expenses where UserId = @Id AND CreditOrDebit = 'D') ,0 );

set @Balance = @CreditAmount - @DebitAmount ;

if(@CreditOrDebit = 'D' AND (@Balance - @Amount) < 0)
begin
select 'you have no sufficient balance ' as Resultmessage, @Balance as balance;
end 

else if @CreditOrDebit = 'D'
begin 
set @Balance = @Balance - @Amount ;
insert into Expenses values(@Comment,@CreditOrDebit,@Amount,GETDATE(),@Balance,@Id ) ;
select 'Added Successfully '  as Resultmessage, @Balance as balance;
end 

else
begin
set @Balance = @Balance + @Amount ;
insert into Expenses values(@Comment,@CreditOrDebit,@Amount,GETDATE(),@Balance,@Id ) ;
select 'Added Successfully' as Resultmessage, @Balance as balance;
end

end
GO
```

# [dbo].[SP_InsertUser]

```sql
/****** Object:  StoredProcedure [dbo].[SP_InsertUser]    Script Date: 09-04-2024 16:38:07 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE proc  [dbo].[SP_InsertUser] 
@Name varchar(100),
@Email varchar(320),
@Mobile varchar(10),
@Address varchar(100),
@Password varchar(100)
as
begin 
declare @Id bigint ;
select @Id = Users.UsersId from Users where Email = @Email ;

if @Id IS NOT NULL
begin 
select ' User Already Exist' as Resultmessage, 0 as ID
return
end 

insert into Users values(@Name,@Email,@Mobile,@Address,GETDATE(),@Password) ;
select 'Register Successfully' as Resultmessage,1 as ID
select * from Users where Email = @Email ;
end
GO
```

# [dbo].[SP_ShowAllExpenses]

```sql
/****** Object:  StoredProcedure [dbo].[SP_ShowAllExpenses]    Script Date: 09-04-2024 16:38:07 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE proc [dbo].[SP_ShowAllExpenses] --'naneradalsukh@gmail.com','Daywise','2024-03-22','2024-03-24',null
@Email varchar(320),
@Flag varchar(100),
@startdate varchar(50),
@enddate varchar(50),
@creditOrDebit varchar(10)
as
begin 

if @Email IS NULL
begin
select 'you are not registered' as ResponseMessage, 0 as ID
return
end

declare @USerId bigint ;
declare @cnt1 bigint ;
declare @cnt2 bigint ;
set @cnt1 = 0 ;
set @cnt2 = 0 ;
select @USerId = UsersId from Users  where Email = @Email 


if @USerId IS NULL
begin
select 'You are not valid user' as ResponseMessage, 0 as ID
return
end

select @cnt1 = count(ExpenseId) from Expenses where UserId = @USerId ;


if @cnt1 = 0 
begin
select 'You do not have any transaction yet' as ResponseMessage, 0 as ID
return
end

if @Flag = 'AllExpense'
begin
select * from Expenses where UserId = @USerId ;
select 'fetched successfully' as ResponseMessage, 1 as ID
end

if @Flag = 'Daywise'
begin
if @creditOrDebit IS NOT NULL
begin
select @cnt2 = count(ExpenseId) from Expenses where UserId = @USerId AND FORMAT(ExpenseTime, 'yyyy-MM-dd') between  @startdate AND @enddate and CreditOrDebit = @creditOrDebit;
end
else
begin
select @cnt2 = count(ExpenseId) from Expenses where UserId = @USerId AND FORMAT(ExpenseTime, 'yyyy-MM-dd')  between  @startdate AND @enddate ;
end
if @cnt2 = 0
begin
select 'You do not  any transaction in this range' as ResponseMessage, 0 as ID
return
end
if @creditOrDebit IS NOT NULL
begin
select * from Expenses where UserId = @USerId AND FORMAT(ExpenseTime, 'yyyy-MM-dd')   between  @startdate AND @enddate and CreditOrDebit = @creditOrDebit;
end
else
begin
select * from Expenses where UserId = @USerId AND FORMAT(ExpenseTime, 'yyyy-MM-dd')  between  @startdate AND @enddate ;
end
select 'fetched successfully' as ResponseMessage, 1 as ID
end
end
GO
```
