# [dbo].[LoginSP]

```sql
/****** Object:  StoredProcedure [dbo].[LoginSP]    Script Date: 09-04-2024 16:38:53 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE proc  [dbo].[LoginSP] 
@Email varchar(320),
@Password varchar(100),
@Flag varchar(100)
as
begin 

if @Flag IS null
begin
select 'Please Provide flag' as ResponseMessage, 0 as ID 
return 
end 

declare  @Id bigint ; 

if @Flag = 'UserLogin'
begin
select @Id = UserTBLId from UserTbl where Email = @Email AND Password = @Password ;  

if @Id is null
begin
select 'provided credetails is wrong' as ResponseMessage , 0 as ID 
return
end
select 'Login successfully' as ResponseMessage, 1 as ID
end

if @Flag = 'AdminLogin'
begin
select @Id = AdminTblId from AdminTbl where Email = @Email AND Password = @Password ;  
if @Id is null
begin
select 'provided credetails is wrong' as ResponseMessage , 0 as ID 
return
end
select 'Login successfully' as ResponseMessage, 1 as ID
end

end 

GO
```

```sql
/****** Object:  StoredProcedure [dbo].[Publish_AddPrice_SP]    Script Date: 09-04-2024 16:38:53 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE proc [dbo].[Publish_AddPrice_SP]
@Price bigint,
@ActivityId bigint,
@EventId bigint,
@Flag varchar(100)
as
begin

if @Flag = 'Publish'
begin

declare @count bigint ;
select @count = COUNT(ActivityId) from ActivityTbl where EventId = @EventId ;

if @count = 0
begin 
select 'You can not publish this becuase you have to first add atleast one activity' as ResponseMessage, 0 as ID
return 
end

update EventTbl set IsPublish =1 where EventId = @EventId ;
select 'Publish successfully' as ResponseMessage, 1 as ID
end 

if @Flag = 'AddPrice'
begin
update ActivityTbl set Price = @Price where ActivityId = @ActivityId ;
select 'Updated Successfully' as ResponseMessage, 1 as ID 
end

end 
GO
```

# [dbo].[Showing_Event_Activity_SP]

```sql
/****** Object:  StoredProcedure [dbo].[Showing_Event_Activity_SP]    Script Date: 09-04-2024 16:38:53 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE proc [dbo].[Showing_Event_Activity_SP] --1,'ActibityShow'
@EventId bigint,
@Flag varchar(1000)

as 
begin 

declare @count bigint ;

if @Flag IS NULL
begin
select 'please provide valid flag' as ResponseMessage, 0 as ID 
return
end

if @Flag = 'AdminEvents'
begin
select @count = COUNT(EventId) from EventTbl where IsPublish = 0 ;
if @count = 0
begin
select 'Do not have any Event To publish' as ResponseMessage, 0 as ID
return 
end

select 'Fetched Successfully' as ResponseMessage, 1 as ID 
select * from EventTbl where IsPublish = 0; 
end

if @Flag = 'EmployeeEvent'

begin
select @count = COUNT(EventId) from EventTbl where IsPublish = 1 ;
if @count = 0
begin
select 'DO not have any scheduled by Admin' as ResponseMessage, 0 as ID
return 
end

select 'Fetched Successfully' as ResponseMessage, 1 as ID 
select * from EventTbl  where IsPublish = 1 ;
end

if @Flag = 'ActibityShow'
begin


select @count = COUNT(ActivityId) from ActivityTbl where EventId = @EventId ;
if @count = 0
begin
select 'Do not have any activity yet' as ResponseMessage, 0 as ID
return 
end

select 'Fetched Successfully' as ResponseMessage, 1 as ID 
select * from ActivityTbl where EventId = @EventId ;
end 

end 




GO
```
