```sql
/****** Object:  StoredProcedure [dbo].[UserBooks]    Script Date: 09-04-2024 16:40:41 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE proc [dbo].[UserBooks]
@UserId int,
@OperationType varchar(50)
as 
begin try

if(@OperationType = 'UserBooks')
begin
select BookBorrowHistory.Bookid, Book.title, Book.Edition from BookBorrowHistory inner join Book on BookBorrowHistory.Bookid = Book.Id where BookBorrowHistory.BorrowBy = @UserId AND BookBorrowHistory.ReturnTime IS NULL;
end;

if(@OperationType = 'AvailableForUser')
begin
select Id,Title,Edition from Book where AvailableBook > 0 AND Book.ID NOT IN (select Bookid from BookBorrowHistory where BorrowBy = @UserId AND BookBorrowHistory.ReturnTime IS NULL);
end;

end try
begin catch
select 'there is an issue'
end catch ;
GO
```
